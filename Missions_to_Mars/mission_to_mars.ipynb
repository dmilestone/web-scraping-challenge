````python
from bs4 import BeautifulSoup as bs
import requests
from splinter import Browser
````

````python
#URL to be scraped
url = 'https://mars.nasa.gov/news/'
response = requests.get(url)
````

````python
# Create BeautifulSoup object; parse with 'html.parser'
soup = bs(response.text, 'html.parser')
````

````python
print(soup.prettify())
````

````python
results = soup.find_all('div', class_="content_title")
results
````

````python
# A blank list to hold the headlines
news_titles = []
# Loop over div elements
for result in results:
    # If td element has an anchor...
    if (result.a):
        # And the anchor has non-blank text...
        if (result.a.text):
            # Append the td to the list
            news_titles.append(result)
news_titles
````

````python
finalnewstitles = []
# Print only the headlines
for x in range(6):
    var=news_titles[x].text
    newvar = var.strip('\n\n')
    finalnewstitles.append(newvar)
finalnewstitles
````

````python
presults = soup.find_all('div', class_="rollover_description_inner")
presults
````

````python
news_p = []
# Print only the headlines
for x in range(6):
    var=presults[x].text
    newvar = var.strip('\n\n')
    news_p.append(newvar)
news_p
````
